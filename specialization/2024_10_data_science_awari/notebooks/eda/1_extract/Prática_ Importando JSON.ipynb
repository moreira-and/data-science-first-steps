# **Importando JSON**

## **Visão geral**

**JSON** é a sigla para *Javascript Object Notation*. É um formato de arquivo (e de dados) estruturado, porém flexível, que é legível tanto a humanos como a máquinas.

O fornecimento e consumo de dados JSON é muito comum na comunicação entre sistemas web e  microsserviços.

## **Objetivo da prática**

Fazer o passo a passo de como importar arquivos JSON (com a extensão ".json") e manipulá-los com Pandas.

## **Realizando a importação**

Você verá que a importação de um arquivo JSOn é muito parecida com a de um arquivo CSV. Na verdade, Pandas converte tudo para DataFrames.

````
# Importando Pandas
import pandas as pd
````

Carregando um arquivo JSON qualquer:

````
# Atribuindo um arquivo ".json" a uma variável
arquivo = "https://raw.githubusercontent.com/persival-iania/json-files/master/Curso%20Pandas/extras/dados/aluguel.json"
````

Agora, usamos o método ```read_json``` (note que é similar ao ```read_csv```) para importar o arquivo para a memória, como DataFrame:

````
# Importando o arquivo JSON que atribuímos à variável arquivo
df = pd.read_json(arquivo)
````

Para ler os dados, basta chamarmos:

````
# Exibe os dados do arquivo
df
````

## **Recursos adicionais**

A função ```read_json``` de Pandas possui uma série de parâmetros adicionais, que nos ajudam a tratar este tipo de arquivo já na sua importação.

A documentação completa está disponível em: https://pandas.pydata.org/docs/reference/api/pandas.read_json.html.

Procure estudá-la e mantê-la por perto enquanto estiver trabalhando com este formato de arquivo.
