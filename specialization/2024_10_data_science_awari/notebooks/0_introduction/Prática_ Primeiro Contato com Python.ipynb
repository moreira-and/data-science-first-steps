### **Comentários**

Usamos **comentários** para anotar o que o código está fazendo. Eles ajudam outras pessoas a entender seu código e também podem ser úteis para você lembrar do que fez depois de um tempo.

### **Variáveis**

Até agora, você usou código para fazer um cálculo e imprimir o resultado, e o resultado não pôde ser reaproveitado.

No entanto, é comum você precisar salvar o resultado para trabalhar com ele mais tarde. Para isso, usamos **variáveis**.

#### **Criando variáveis**

A próxima célula de código cria uma variável chamada `test_var` e atribui a ela o valor que obtemos quando adicionamos 5 a 4.

Em seguida, imprimimos o valor atribuído à variável, que é 9.

````python
# Criando variável chamada test_var e atribuindo a ela uma soma
soma = 4 + 7

# Imprimindo o valor de test_var
print(soma)
````

Em geral, para trabalhar com uma variável, você precisa começar selecionando o nome que deseja usar.

Nomes de variáveis ​​são idealmente curtos e descritivos. Eles também precisam atender a vários requisitos:

- Eles não podem ter espaços (por exemplo, `test var` não é permitido);
- Eles podem incluir apenas letras, números e sublinhados (por exemplo, `test_var!` não é permitido);
- Eles devem começar com uma letra ou sublinhado (por exemplo, `1_var` não é permitido);
Então, para criar a variável, você precisa usar `=` para atribuir o valor que você deseja que ela tenha;

Com o tempo, você aprenderá a selecionar bons nomes para variáveis ​​Python. Não há problema em se sentir desconfortável agora. A melhor maneira de aprender é visualizando e praticando.

#### **Manipulando variáveis**

É possível alterar o valor atribuído a uma variável.

Na célula de código abaixo, alteramos o valor de `my_var` de 3 para 100.

````python
# Defina o valor de uma nova variável para 3
my_var = 3

# Imprima o valor atribuído a my_var
print(my_var)

# Altere o valor da variável para 100
my_var = 100

# Imprima o novo valor atribuído a my_var
print(my_var)
````

Observe que, em geral, sempre que você define uma variável em uma célula de código, todas as células de código na sequêncua também têm acesso à variável.

Por exemplo, usamos a próxima célula de código para acessar os valores de `my_var` (da célula de código acima) e `soma` (criada anteriormente neste tutorial).

````python
print(my_var)
print(soma)
````

Se quisermos aumentar o valor de ```my_var```, que é 100, em mais 3 unidades, podemos fazer:

````python
# Aumentando my_var
my_var = my_var + 3

# Imprimindo o valor atribuído a my_var
print(my_var)
````

#### **Usando várias variáveis**

É comum que um programa use diversas variáveis. Isso é especialmente útil quando temos que fazer um cálculo longo, com várias entradas.

Na próxima célula de código, calculamos o número de segundos em quatro anos. Este cálculo usa cinco entradas.

````python
# Criando variáveis
num_years = 4
days_per_year = 365
hours_per_day = 24
mins_per_hour = 60
secs_per_min = 60

# Calcular o número de segundos em quatro anos
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
````

Conforme calculado acima, há 126144000 segundos em quatro anos.

Note que é *possível* fazer esse cálculo sem variáveis ​​como apenas `60 * 60 * 24 * 365 * 4`, mas é muito mais difícil verificar se o cálculo sem variáveis ​​não tem algum erro, pois não é tão legível.

Quando usamos variáveis ​​(como `num_years`, `days_per_year`, etc), podemos acompanhar melhor cada parte do cálculo e verificar e corrigir erros com mais facilidade.

Observe que é particularmente útil usar variáveis ​​quando os valores das entradas podem mudar.

Por exemplo, digamos que queremos melhorar um pouco nossa estimativa atualizando o valor do número de dias em um ano de 365 para 365,25, para contabilizar os anos bissextos. Então podemos alterar o valor atribuído a `days_per_year` sem alterar nenhuma das outras variáveis ​​e refazer o cálculo.

````python
# Atualização para incluir anos bissextos
days_per_year = 365.25

# Calcular o número de segundos em quatro anos
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
````

**Importante!** Você deve ter notado que escrevemos 365,25 (com vírgula) no texto, mas digitamos `365.25` (com ponto) no código. Em Python, números decimais são separados por ponto final (`.`) e não vírgula (`,`). Você aprenderá mais decimais e partes fracionárias de números ao ver sobre tipos de dados.

### **Depuração (*debugging*)**

Um erro comum ao trabalhar com variáveis ​​é introduzir erros de digitação acidentalmente.

Por exemplo, se escrevermos `hours_per_day` como `hours_per_dy` (**dy** em vez de **day**), o Python apresentará um erro.

Rode a célula abaixo e veja o que ocorre:

````python
print(hours_per_dy)
````

Quando você vê `NameError`, é uma indicação de que deve verificar como escreveu a variável a que o erro se refere. Então, para corrigir o erro, você só precisa corrigir a grafia, como na célula abaixo:

````python
print(hours_per_day)
````
