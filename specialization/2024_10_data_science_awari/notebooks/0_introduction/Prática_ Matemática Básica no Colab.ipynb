### **3) Calculando uma média aritmética**

Há duas fórmulas a seguir (a primeira, que pode parecer enigmática a quem não é familiarizado). Porém, ela envolve operações muito simples.

Sabendo que a fórmula da média aritmética é:

\begin{align}
       MA = \frac{1}{n} \sum_{i=1}^n a_i
    \end{align}

Ou, de forma simplificada:

\begin{align}
       MA = \frac{a_1 + a_2 + a_2 + ... + a_n}{n}
    \end{align}

a) Como você montaria o cálculo para chegar à média aritmética para o conjunto: {$2, 14.3, 4.6, -2.4, 7.5, 3.2, 6$}?

b) Tente fazer isso por conta própria em Python. Pense um pouco e veja até onde consegue chegar.

c) Agora, dê uma olhada na biblioteca nativa [statistics](https://docs.python.org/3/library/statistics.html). Qual função dela você usaria? Saberia construí-la em uma célula aqui no Colab?

d) Arredondando o número final para duas casas decimais, como o valor ficaria?

````python
# Tente resolver a questão 3 aqui

import numpy as np
ai = np.array([2,14.3,4.6,-2.4,7.5,3.2,6])
n=len(ai)
Asum=0

for i in range(n):
  Asum += ai[i]

Aavg=Asum/(n)

Aavg
````

````python
# Tente resolver usando a biblioteca "statistics"
dataArray = np.array([2,14.3,4.6,-2.4,7.5,3.2,6])
sum(dataArray)/(len(dataArray))
````

````python
# Tente resolver arredondando o resultado para duas casas decimais
dataArray = np.array([2,14.3,4.6,-2.4,7.5,3.2,6])
round(sum(dataArray)/(len(dataArray)),2)
````

````python
from statistics import mean
dataArray = np.array([2,14.3,4.6,-2.4,7.5,3.2,6])
round(mean(dataArray),2)
````

Você deve ter visto essa letra grega (Sigma) em alguma lugar nesta aula. Tente avaliar esta expressão:

````python
# Tente resolver a questão 6 aqui
import numpy as np
n=5
Asum=0
ArraySum = np.array(0)

for i in range(n):
  Asum += 2*(i+1)
  ArraySum=np.append(ArraySum,2*(i+1))

print(ArraySum[1:])
Asum
````
