Vamos brincar um pouco com a biblioteca. É muito comum usarmos Matplotlib em conjunto com outras bibliotecas numéricas, principalmente Numpy e Pandas.

```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

### **Gráfico de linha**

- Um dos gráficos mais comuns para analisarmos os dados ao longo do tempo.
- Analisamos o comportamento de uma varíavel no eixo X em relação ao eixo Y.
- O produto final é uma linha plotada no gráfico, conectando o par (x, y).

```python
# Gerar uma par ordenado de x e y de forma randômica, somente como exemplo
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

# Plotar os valores
plt.plot(x, y)  # simplesmente utilizamos a função plot e passamos o par de coordenadas x e y
plt.title("Exemplo") # incluir um título no gráfico
plt.xlabel("Eixo x") # nomear o eixo x
plt.ylabel("Eixo y") # nomear o eixo y

plt.show() # comando que efetivamente plota o gráfico; os comandos anteriores são para os parâmetros do mesmo
```

### **Grafico de barras**

- Muito utilizados para quantificar categorias ou classes de dados.
- As quantidades ficam no eixo 7.
- As categorias no eixo x.
- O gráfico mostra a distribuição dos valores de cada categoria nas barras.

```python
np.random.seed(42)
```

```python
# Gerar valores aleatórios como exemplo
np.random.seed(42)
x = np.arange(14)
y = np.arange(14)

# Plotar os valores
plt.bar(x, y) # desta vez, utilizaremos a função bar que corresponde a um gráfico de barras
plt.title("Exemplo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

### **Histograma**

- Também conhecido como distribuição de frequências.
- É a representação gráfica em colunas ou em barras (retângulos) de um conjunto de dados previamente tabulado e dividido em classes uniformes ou não uniformes.
- A base de cada retângulo representa uma classe.
- A altura de cada retângulo representa a quantidade ou a frequência absoluta com que o valor da classe ocorre no conjunto de dados para classes uniformes ou a densidade de frequência para classes não uniformes.
- Importante ferramenta da estatística, o histograma também é uma das chamadas "[sete ferramentas da qualidade](https://pt.wikipedia.org/wiki/Sete_ferramentas_da_qualidade)".

```python
# Gerando um array com uma amostra de dados
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(a, bins = [0,20,40,60,80,100]) # desta vez, usaremos a função hist para plotar um histograma; bins significam os espaços de cada barra, onde os dados serão distribuídos
plt.title("histogram")
plt.show()
```

### **Gráfico de setores**

- Também conhecido como gráfico de "pizza" ou "torta".
- Distribui as categorias (ou labels) em um círculo.
- Os valores são distribuídos propocionalmente dentro do círculo.
- Embora seja largamente utilizado, não recomendamos utilizar este gráfico para muitas categorias, pois dificulta a leitura.
- Para muitas categorias, o gráfico de barras é mais indicado.[texto do link](https://)

```python
# Gerando amostras de dados
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels) # a função agora é pie, que gera o gráfico de "pizza" ou "torta"; passamos como parâmetro o array e as categorias ou labels do gráfico
plt.title("Exemplo")

plt.show()
```

### **Boxplot**

- Em português, "diagrama de caixa".
- Visualização muito utilizada na análise exploratória de dados.
- Apresenta as principais métricas estatísticas de centralidade de dados.
- Basicamente, Boxplot apresenta de forma visual métricas estatísticas da variável como valor mínimo, máximo, os percentis, assim como a mediana e os valores outliers acima e abaixo do mínimo e máximo.
- Todas estas métricas serão bastante exploradas no módelo de estatística descritivas.

```python
# Gerando um array com dados aleatórios
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7)) # aqui, podemos configurar o tamanho também

plt.boxplot(data) # neste caso, vamos usar a função boxplot
plt.title('Exemplo')

plt.show()
```

### **Scatter plot**

- Em português, "gráfico de dispersão".
- Descrito por Francis Galton, o *scatter plot* é usado para verificar se existe uma relação entre causa e efeito entre duas variáveis numéricas.
- São representações gráficas do relacionamento entre duas variáveis numéricas.
- O *scatter plot* utiliza pontos para representar essa relação. Cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical.

```python
# Vamos gerar 2 arrays com valores para serem a base do gráfico
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y) #aqui, usamos a função scatter, que significa Scatter Plot ou Gráfico de Dispersão
plt.title('Exemplo')
plt.show()
```
